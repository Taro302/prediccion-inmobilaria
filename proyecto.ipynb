```python
categorical_cols = ['DISTRITO']
df_filtrado = pd.get_dummies(df_filtrado, columns=categorical_cols)
#df_filtrado=df_filtrado.drop('HABITACIONES',axis=1)
df_filtrado=df_filtrado.drop('ANTIGÜEDAD(AÑOS)',axis=1)
df_filtrado
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.neural_network import MLPRegressor

# Obtener las columnas codificadas generadas por pd.get_dummies
new_columns = list(df_filtrado.columns)
label = 'PRECIO VENTA(USD)'
new_columns.remove(label)

# Actualizar la lista features
features = [] + new_columns

# Preparemos las variables
X = df_filtrado[features]
y = df_filtrado[label]

# Separemos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=17, test_size=0.25)
print(len(X_train), len(X_test))
print(len(y_train), len(y_test))

# Modelos para variable dependiente conocida: https://scikit-learn.org/stable/supervised_learning.html
models = [
    ("GradientBoostingRegressor", GradientBoostingRegressor()),
    ("SVR", SVR()),
    ("DecisionTreeRegressor", DecisionTreeRegressor()),
    ("RandomForestRegressor", RandomForestRegressor()),
    ("MLPRegressor", MLPRegressor()),
    ("LinearRegression", LinearRegression())
]

results = []
r2 = []
ms = []
results.append(np.array(y_test))

for model_name, model in models:
    # Estimar modelo
    model.fit(X_train, y_train)
    
    for dataset in ["train", "test"]:
        y_pred = model.predict(X_train if dataset == "train" else X_test)
        r_sq = model.score(X_train if dataset == "train" else X_test, y_train if dataset == "train" else y_test)
        mse = mean_squared_error(y_train if dataset == "train" else y_test, y_pred)
        
        print(f"Para {model_name} en {dataset}, el R cuadrado es: {r_sq}")
        print(f"Para {model_name} en {dataset}, el MSE es: {mse}")
        
        if dataset == "test":
            r2.append(r_sq)
            ms.append(mse)
```

```python
best_model_index = np.argmax(r2)
best_model_name = models[best_model_index][0]
best_r2 = r2[best_model_index]
best_mse = ms[best_model_index]

X_prueba = {
    'TAMAÑO(M2)': [180],
    'HABITACIONES':[4],
    'BAÑOS': [3],
    'ESTACIONAMIENTO': [1],
    'PISCINA': [0],
    'DISTRITO_Ate Vitarte': [0],
    'DISTRITO_La Molina': [0],
    'DISTRITO_Los Olivos': [0],
    'DISTRITO_San Isidro': [0],
    'DISTRITO_San Miguel': [1]  
}

X_prueba_df = pd.DataFrame(X_prueba)
best_model = models[best_model_index][1]
columnas_entrenamiento = list(X.columns)
# Realiza la predicción utilizando el modelo cargado
y_pred_prueba = best_model.predict(X_prueba_df)
print(f"El mejor modelo es: {best_model_name}")
print(f"R cuadrado del mejor modelo: {best_r2}")
print(f"MSE del mejor modelo: {best_mse}")
print("La predicción de Precio_venta(USD) para el ejemplo es:", y_pred_prueba)
```
